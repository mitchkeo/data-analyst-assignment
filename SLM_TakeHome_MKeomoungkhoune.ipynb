**Question a)** How many households are in the postal routes that have a median household income greater than $80,000 and a median household age of 40-55?

```python
#filtering median incomes greater than $80,000 and median household ages between 40 and 55
Households = data1A[(data1A['MedianIncome'] > 80000) & (data1A['Median Age'] >= 40) & (data1A['Median Age'] <=55)]
Households.head()
```

```python
#calculating total households from filter: 59,869 Households 
Total_Households = Households['TotalHouseholds']
print("Number of households with median household income greater than $80,000 and median hosuehold age of 40-55:")
print("----------------------------------------------------------------------------------------------------------")
print(Total_Households.sum(),"Households")
```

**Question b)** Let's say you want to target 100K households in Atlanta for our SLM shared mail program. Which Geocodes do you choose and why?

```python
#targeting 100k households 
Geocodes = data1A[(data1A['MedianIncome'] > 55000) & (data1A['CityName'] == 'ATLANTA') & (data1A['Median Age'] < 44) & (data1A['PercentHouseholdWithChildren'] <= 28)]
Geocodes.head()
```

```python
#targeted geocodes
print(Geocodes['GeoCode'].describe())
print('\n')
print("Geocodes to target:")
print("-------------------")
print(Geocodes['GeoCode'])
```

```python
#Determining total number of households and geocodes associated with 100k household target for share mail program in Atlanta
print("Number of households for 100k household target in Atlanta for shared mail program:")
print("----------------------------------------------------------------------------------")
Geo_Households = Geocodes['TotalHouseholds']
print(Geo_Households.sum(),"Households")
print('\n')
print("Number of geocodes for 100k household target in Atlanta for shared mail program:")
print("--------------------------------------------------------------------------------")
print(len(pd.value_counts(Geocodes['GeoCode'])),"Geocodes")
```

**Question d)** Dataset 1C is a list of Geocodes that have been previously mailed. Using your answer from question 2, what % of this market has been mailed before? What % has been mailed 2 times?

```python
#merge datas1A, dataset1B, and dataset1C
data1A_1B_1C = pd.merge(data1A_1B,data1C, on = "GeoCode")
data1A_1B_1C.head()
```

```python
twice_mailed = data1A_1B_1C.pivot_table(columns = ["GeoCode"], aggfunc = 'size')
twice_mailed
```

```python
#determing percent of  market that has been mailed twice before: dividing number of geocodes mailed twice before by total geocodes
twice_mailed = data1A_1B_1C.pivot_table(columns = ["GeoCode"], aggfunc = 'size')
print("Percent of market that has been mailed twice before: %.2f" % ((((twice_mailed == 2).sum())/len(data1A))*100),"%")
```

```python
#determining percent of market that has been mailed before: dividing number of geocodes mailed before by total geocodes 
print("Percent of market that has been mailed before: %.2f" % float(len(twice_mailed)/len(data1A)*100),"%|")
```

```python
mailed_twice_before = (((twice_mailed == 2).sum())/len(data1A))*100
mailed_before = float(len(twice_mailed)/len(data1A)*100)
mailed = [mailed_twice_before,mailed_before]
x = ['mailed twice before','mailed before']
plt.pie(mailed, labels = x, autopct = '%2.1f%%')
plt.figsize = (10,10)
plt.show
```

```python
explode = (0.62,0.39)
fig, ax = plt.subplots()
ax.pie(mailed,
       explode = explode,
       labels = x,
       autopct = '%1.1f%%',
       shadow = True,
       startangle = 90)
ax.axis('equal')
ax.set_title('Percentage of Geocodes previously mailed and previously mailed twice') 
```

*dataset 2A cleaning and manipulation*

```python
#check for missing/null values and data type
data2A.info()
```

```python
#changing line item value dollar values to numeric values
data2A['Line Item Value'] = pd.to_numeric(data2A['Line Item Value'].replace('[^0-9\.-]','', regex = True))
data2A
```

```python
data2A.info()
```

```python
#grouping dataset2A by Order ID and calculating total cutomer cost by Order ID
ID = data2A.groupby('Order ID')
for group, group_df in ID:
    print("Order ID:",group)
    print(group_df)
    print("Total customer cost for Order ID: $",group_df['Line Item Value'].sum())
    print("\n")
```

```python
#designating each column to grouped IDs
ID1 = data2A.groupby(['Order ID'])['Line Item Value'].sum() 
ID2 = data2A.groupby(['Order ID'])['Order Date'].max()
ID3 = data2A.groupby(['Order ID'])['Campaign Cell'].max()
```

```python
#merging each column by grouped IDs 
data2A_new = pd.merge(ID1,ID2, on = "Order ID")
dataset2A = pd.merge(data2A_new,ID3, on = "Order ID")
dataset2A.to_csv(r'dataset2A_clean')
dataset2A.head()
```
